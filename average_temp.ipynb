```python
import pandas as pd
import kaggle 
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sb
```

```python
avg_temp = pd.read_csv('Average Temperature 1900-2023.csv')
```

```python
avg_temp
```

```python
# Set the style of seaborn
sb.set(style="whitegrid")

# Create a line plot
pt.figure(figsize=(20, 6))
sb.lineplot(x='Year', y='Average_Fahrenheit_Temperature', data=avg_temp, marker='o')

# Set labels and title
pt.title('Temperature over the Years')
pt.xlabel('Year')
pt.ylabel('Average_Fahrenheit_Temperature')

# Show the plot
pt.show()
```

```python
# filtering first 60 years
df_60 =  avg_temp[avg_temp['Year'] <= avg_temp['Year'].min() + 60]

# Set the style of seaborn
sb.set(style="whitegrid")

# Create a line plot
pt.figure(figsize=(20, 6))
sb.lineplot(x='Year', y='Average_Fahrenheit_Temperature', data=df_60, marker='o')

# Set labels and title
pt.title('Temperature over the Years')
pt.xlabel('Year')
pt.ylabel('Average_Fahrenheit_Temperature')

# Show the plot
pt.show()
```

```python
# filtering first 60 years
df_last =  avg_temp.tail(60)

# Set the style of seaborn
sb.set(style="whitegrid")

# Create a line plot
pt.figure(figsize=(20, 6))
sb.lineplot(x='Year', y='Average_Fahrenheit_Temperature', data=df_last, marker='o', color = 'red')

# Set labels and title
pt.title('Temperature over the Years')
pt.xlabel('Year')
pt.ylabel('Average_Fahrenheit_Temperature')

# Show the plot
pt.show()
```

```python
# Convert Fahrenheit to Celsius
avg_temp['Temperature_Celsius'] = (avg_temp['Average_Fahrenheit_Temperature'] - 32) * 5/9

# Display the DataFrame with both Fahrenheit and Celsius temperatures
print(avg_temp)
```
